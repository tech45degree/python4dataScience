## Querying a dataframe

Querying a database means finding some values based on certain conditions. For example you want to find out the albums having rating greater and equal to 9

```python
dfalbum
```

```python
dfalbum[dfalbum['Rating']>=9]
```

Notice that in the above result all the columns are displayed. If you want to access a specific column, then you can use **loc** for that purpose.

```python
dfalbum.loc[dfalbum['Rating']>=9,['Artist','Album','Rating']]
```

# Grouping

Grouping works similar to "group by" as in databases. Pandas also provides a "group by" function which serves the same purpose.

```python
df_cars.groupby('fuel-type')['price'].mean()
```

```python
df_cars.groupby('num-of-doors')['price'].mean()
```

```python
df_fuelType = df_cars.groupby('fuel-type')['price'].mean()
```

```python
df_fuelType
```

```python
df_fuelType = df_cars.groupby('fuel-type')['price'].mean().reset_index()
```

```python
df_fuelType
```

```python
type(df_fuelType)
```

```python
num_of_doors = df_cars.groupby('num-of-doors')['price'].mean().reset_index()
```

```python
type(num_of_doors)
```

# Pivoting and Melting

**Pivoting**

```python
pd.pivot_table(group,index='fuel-type',columns='num-of-doors',values = 'price',aggfunc='mean',fill_value=0)
```

```python
pivoted = pd.pivot_table(group,index='fuel-type',columns='num-of-doors',values = 'price',aggfunc='mean',fill_value=0)
```

```python
pivoted
```

```python
x = pivoted.reset_index()
```

```python
x
```

```python
type(x)
```

**Melting**

```python
melted = pd.melt(x,id_vars='fuel-type',value_name='Mean Value')
```

```python
melted
```
